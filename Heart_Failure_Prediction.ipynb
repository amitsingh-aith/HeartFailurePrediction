```python
# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=np.number)

# Calculate and display the correlation matrix
correlation_matrix = numeric_df.corr()
display(correlation_matrix)

#And here is the heat mapp for the same.
sns.heatmap(correlation_matrix)
```

Let's Use joint plot for correlation.

```python
sns.jointplot(x= 'Age', y= 'MaxHR', data= df, kind= 'hex')
```

As we saw the correlation via jointplot, now we use a reggression technique to find the best fit line, for further analysis.

```python
sns.jointplot(x= 'Age', y= 'MaxHR', data= df, kind= 'reg')
```

```python
sns.jointplot(x= 'Cholesterol', y= 'MaxHR', data= df, kind= 'reg')
```
